```python
import numpy as np
import pandas as pd
```

```python
food=pd.DataFrame([['tofu',90,'portland','american','sweet'],
                   ['pizza',150,'sinjuku','japanese','sour'],
                   ['sandwich',40,'meyongdong','korea','salty'],
                   ['bread',80,'pattaya','Thai','sweet'],
                   ['tokoyaki',100,'Mumbai','japanese','savory'],
                   ['bottle ice cream',40,'Delhi','Thai','creamy'],
                   ['dhokla',70,'pune','Indian','spicy'],
                   ['salad',200,'Kolkata','japanese','spicy'],
                   ['mac&chezz',50,'pune','Spanish','spicy'],
                   ['malai kofta',70,'Hyderbad','Spanish','spicy'],
                   ['dal bati',230,'jaisalmer','rajasthani','spicy'],
                   ['red sauce pasta',20,'Naples','Italian','salty']
                  ]
                  ,columns=['Name','Price','Place','Cuisine','Taste'])
food
```

```python
for x in food.index:
    if food.loc[x,'Price']==food['Price'].max():
         print('Costliest Dish Name = ',food.loc[x,'Name'],'\nDish Price =',food.loc[x,'Price'])
```

```python
food['Rating']=pd.Series([4.4,5,1.3,5,3,3.5,2.4,2.2,5,4.4,1.4,5],index=range(0,12))
food
```

```python
for x in food.index:
    if food.loc[x,'Price']==food["Price"].min():
         print('Cheapest Dish = ',food.loc[x,'Name'],'\nDish Price = ',food.loc[x,'Price'])
```

```python

for x in food.index:
    if food.loc[x,'Rating']==food["Rating"].max():
          print('Most Popular Cuisine = ',food.loc[x,'Cuisine'],'\nDish Rating = ',food.loc[x,'Rating'])
```
